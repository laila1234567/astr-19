```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.normal(0, 1, 1000)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(random_numbers, bins=100, color='lightcoral', edgecolor='black')
plt.xlabel('Random Number Value')
plt.ylabel('Frequency')

plt.title('Histogram of Normally Distributed Random Numbers')
plt.savefig('histogram_normal_distribution.pdf')
plt.show()
```
